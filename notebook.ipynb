#### Visualisasi Berdasarkan Status Mahasiswa

```python
# Fungsi untuk Visualisasi
def distribusi_berdasarkan_status(df, column, figsize=(10, 5), title=None):
    plt.figure(figsize=figsize)
    sns.countplot(data=df, x=column, hue='Status', palette='viridis')
    plt.xlabel(column)
    plt.ylabel('Jumlah Mahasiswa')
    if title is None:
        title = f'Distribusi Status Mahasiswa berdasarkan {column}'
    plt.title(title, fontweight='bold', fontsize=14)
    plt.xticks(rotation=45, ha='right') 
    plt.legend(title='Status') 
    plt.tight_layout()
    plt.show()
```

```python
# Kategorikal
feature_categorical = ['Marital_status', 'Gender']
for col in feature_categorical:
    distribusi_berdasarkan_status(df, col)
```

## Data Preparation / Preprocessing

```python
# --- Pengecekan Missing Values dan Duplikasi ---
print("Duplikasi:", df.duplicated().sum())
print("Missing values:\n", df.isnull().sum())
```

Pada tahap ini saya melakukan pengecekan duplikasi data dan missing values karena langkah ini sangat penting untuk memastikan kualitas data yang akan dianalisis. Data yang tidak bersih dapat menyebabkan bias atau kesalahan dalam proses modeling.

```python
# --- Drop Kolom yang Tidak Relevan ---
deleted_columns = [
    "Unemployment_rate", "GDP", "Inflation_rate",
    "Curricular_units_2nd_sem_evaluations", "Curricular_units_2nd_sem_without_evaluations",
    "Curricular_units_1st_sem_without_evaluations", "Curricular_units_1st_sem_evaluations",
    "Educational_special_needs", "International", "Course", "Application_mode",
    "Nacionality", "Mothers_qualification", "Fathers_qualification",
    "Mothers_occupation", "Fathers_occupation", "Marital_status",
    "Age_at_enrollment", "Previous_qualification"
]
df.drop(columns=deleted_columns, inplace=True)
```

Kolom ini dihapus karena dianggap tidak relevan, redundan, memiliki banyak nilai unik, atau bersifat umum. Tujuannya untuk menyederhanakan fitur dan menghindari noise.

```python
# --- Label Encode Target ---
le = LabelEncoder()
df['Status'] = le.fit_transform(df['Status'])  
```

Label encoding diperlukan untuk target karena model hanya bisa memproses angka. 

```python
# --- Preprocessing ---
categorical_cols = ['Daytime_evening_attendance', 'Displaced',
                    'Debtor', 'Tuition_fees_up_to_date', 'Gender', 'Scholarship_holder']
numerical_cols = [col for col in df.columns if col not in categorical_cols + ['Status']]
```

```python
# One-Hot Encoding
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Standardisasi
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

One-hot encoding untuk fitur kategorikal agar model memahami perbedaan tiap kategori. StandardScaler menyesuaikan skala fitur numerik agar setara.

```python
# --- Mapping Label ke Binary ---
df['Status'] = df['Status'].replace({1: 1, 2: 1})  # Enrolled & Graduate jadi 1

# --- Split Data ---
X = df.drop('Status', axis=1)
y = df['Status']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)

# --- SMOTE (untuk keseimbangan kelas) ---
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

SMOTE digunakan karena kelas target tidak seimbang, seperti yang terlihat pada grafik distribusi sebelumnya. Tanpa ini, model bisa bias terhadap kelas mayoritas.

## Modeling

Tahapan ini bertujuan untuk menemukan kombinasi hyperparameter terbaik untuk model XGBoost, sehingga model dapat memberikan kinerja optimal dalam memprediksi data, terutama setelah penanganan data imbalance

```python
# --- Model XGBoost dengan GridSearch ---
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

grid_search = GridSearchCV(xgb, param_grid, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(X_train_res, y_train_res)
```

## Evaluation

Pada tahap ini, dilakukan evaluasi akhir terhadap kinerja model XGBoost yang telah dilatih dan dioptimalkan melalui GridSearchCV. Langkah ini sangat krusial untuk memahami seberapa baik model dapat membuat prediksi pada data yang belum pernah dilihat sebelumnya (data pengujian), dan untuk memastikan bahwa model siap digunakan untuk memecahkan masalah klasifikasi.

```python
# --- Evaluasi ---
y_pred = grid_search.predict(X_test)

print("Best Parameters:", grid_search.best_params_)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

Pada classification report, dapat dilihat bahwa model XGBoost memiliki kinerja yang cukup baik, dengan akurasi keseluruhan 87%. Penting untuk dicatat bahwa meskipun recall untuk kelas 0 sedikit lebih rendah (0.76) dibandingkan kelas 1 (0.93), ini adalah hal yang umum dalam kasus data tidak seimbang, dan kinerja keseluruhan yang ditunjukkan oleh weighted avg dan akurasi 0.87 menunjukkan model yang solid.

```python
# Hitung confusion matrix
cm = confusion_matrix(y_test, y_pred)
labels = ['Dropout', 'Tidak Dropout']

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

Dari matriks ini, kita bisa melihat secara langsung di mana model bekerja dengan baik dan di mana ia melakukan kesalahan. Misalnya, model jauh lebih baik dalam memprediksi siswa yang Tidak Drop out (558 TN dan hanya 43 FN) dibandingkan dengan memprediksi siswa yang Drop out (216 TP dan 68 FP). Ini membantu memvalidasi metrik dari Classification Report dan memberikan pemahaman visual yang lebih intuitif tentang sebaran kesalahan prediksi model.

## Feature Importance

Pada tahapan ini, dilakukan analisis Feature Importance dari model XGBoost yang sudah dilatih. Langkah ini sangat penting dalam interpretasi model dan data discovery. Model berbasis pohon seperti XGBoost secara internal menghitung seberapa besar kontribusi setiap fitur dalam membuat keputusan prediksi. Dengan memvisualisasikan kepentingan fitur, kita bisa memahami variabel-variabel mana yang paling berpengaruh dalam menentukan apakah seorang siswa akan dropout atau tidak. Informasi ini tidak hanya penting untuk memahami model itu sendiri, tetapi juga bisa memberikan wawasan berharga bagi pembuat kebijakan atau pihak terkait untuk fokus pada faktor-faktor kunci.

Berdasarkan grafik ini, dapat disimpulkan bahwa Curricular_units_2nd_sem_approved adalah faktor paling signifikan yang memengaruhi keputusan dropout. Informasi ini sangat berharga untuk memahami fenomena dropout dan dapat dijadikan dasar untuk mengembangkan intervensi atau kebijakan yang lebih terarah, dengan memfokuskan perhatian pada fitur-fitur yang paling berpengaruh ini.
